```python
import matplotlib.pyplot as plt
from graph import create_random_graph, has_cycle
```

```python
number_of_nodes = 100
runs = 500
```

```python
max_edges = number_of_nodes
step = 1
edges_list = list(range(0, max_edges + 1, step))
if edges_list[-1] != max_edges:
    edges_list.append(max_edges)
```

```python
data = []

for edges in edges_list:
    cycle_count = 0

    for _ in range(runs):
        graph = create_random_graph(number_of_nodes, edges)
        if has_cycle(graph):
            cycle_count += 1
    data.append(cycle_count / runs)
```

```python
plt.plot(edges_list, data, label='Cycle Detection')
plt.title(f"Experiment 1: Edges vs Cycle Detection (n={number_of_nodes})")
plt.xlabel('Number of edges')
plt.ylabel('Probability of cycle')
plt.legend()
plt.grid(True)
plt.savefig("experiment1_graph.png", dpi=300)
plt.show()
```
