```python
import matplotlib.pyplot as plt
import random
import itertools
import math
from graph import (
    Graph, create_random_graph,
    has_cycle, is_connected, is_vertex_cover, MVC,
    BFS, DFS, BFS2, DFS2, BFS3, DFS3
)
```

```python
number_of_nodes = 100
runs = 500

max_edges = int(number_of_nodes * math.log(number_of_nodes))
step = max_edges // 40
edges_list = list(range(0, max_edges + 1, step))
if edges_list[-1] != max_edges:
    edges_list.append(max_edges)
```

```python
data = []

for edges in edges_list:
    cycle_count = 0

    for _ in range(runs):
        graph = create_random_graph(number_of_nodes, edges)
        if has_cycle(graph):
            cycle_count += 1
    data.append(cycle_count / runs)
```

```python
data_connected = []

for edges in edges_list:
    connected_count = 0

    for _ in range(runs):
        graph = create_random_graph(number_of_nodes, edges)
        if is_connected(graph):
            connected_count += 1
    data_connected.append(connected_count / runs)
```

```python
plt.plot(edges_list, data_connected, color='orange', label='Connectivity')
plt.title(f"Experiment 2: Edges vs Connectivity (n={number_of_nodes})")
plt.xlabel('Number of edges')
plt.ylabel('Probability of connectivity')
plt.legend()
plt.grid(True)
plt.savefig("experiment2_graph.png", dpi=300)
plt.show()
```

```python
plt.plot(edges_list, data, label='Has Cycle', color='steelblue')
plt.plot(edges_list, data_connected, label='Is Connected', color='darkorange')
plt.title(f"Experiment 1 and 2 (n={number_of_nodes})")
plt.xlabel('Number of edges')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)
plt.savefig("experiment_combined.png", dpi=300)
plt.show()
```
