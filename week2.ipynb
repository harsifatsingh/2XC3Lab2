```python
import matplotlib.pyplot as plt
import random
import graph as gt
```

```python
def run_vc_experiment(n=8, edge_counts=range(5, 25), runs=20):
    avg_ratios = []

    for j in edge_counts:
        ratios = []

        for _ in range(runs):
            G = gt.create_random_graph(n, j)

            optimal = gt.MVC(G)
            approx = gt.approx_vertex_cover(G)

            if len(optimal) > 0:
                ratios.append(len(approx) / len(optimal))

        avg_ratios.append(sum(ratios) / len(ratios))

    plt.figure()
    plt.plot(list(edge_counts), avg_ratios, marker='o')
    plt.xlabel("Number of edges")
    plt.ylabel("Approximation ratio")
    plt.title("Vertex Cover Approximation Quality")
    plt.grid(True)
    plt.savefig()
    plt.show()

    return avg_ratios
```

```python
run_vc_experiment()
```

```python
def verify_is_vc_relationship(n=10, edges=15, runs=10):
    for i in range(runs):
        G = gt.create_random_graph(n, edges)

        mvc = gt.MVC(G)
        independent_set = [v for v in G.adj if v not in mvc]

        print(f"Run {i+1}:")
        print(" |MVC| =", len(mvc))
        print(" |IS|  =", len(independent_set))
        print(" Sum   =", len(mvc) + len(independent_set))
        print(" Nodes =", G.number_of_nodes())
        print("-" * 30)

verify_is_vc_relationship()
```
